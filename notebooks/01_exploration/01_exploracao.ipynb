```python
import os
from dotenv import load_dotenv 
# Carregar variáveis de ambiente do arquivo .env
load_dotenv(os.path.join('config', '.env'))
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Configurações de conexão
HOST = os.getenv('DB_HOST')
USER = os.getenv('DB_USER')
PASSWORD = os.getenv('DB_PASSWORD')
DATABASE = os.getenv('NEW_DBNAME')
PORT = os.getenv('DB_PORT')  # Porta padrão do PostgreSQL

# String de conexão
CONNECTION_STRING = f'postgresql+psycopg2://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}'

# Criação do engine
engine = create_engine(CONNECTION_STRING)
```

```python
# Leitura dos dados para um DataFrame Pandas
df_customers = pd.read_sql(sql='SELECT * FROM olist_customers', con=engine)
df_customers.head()
```

```python
# Leitura dos dados para um DataFrame Pandas
df_order_items = pd.read_sql(sql='SELECT * FROM olist_order_items', con=engine)
df_order_items.head()
```
